```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
N = 80
P = 120
B = 8

DN = int(N * N / B)
DP = int(P * P / B)
```

```python
img = cv2.imread('./maudy.png', 0)
img = cv2.resize(img, (N,N))
imvec = np.reshape(np.array(img), (N*N,1))

def imshow(imgdata):
    plt.imshow(imgdata, cmap = 'gray')
    plt.show()

imshow(img)
```

```python
def V(matN, matM, matI, matK):
    matN32 = np.float32(matN)
    matM32 = np.float32(matM)
    matI32 = np.float32(matI)
    matK32 = np.float32(matK)
    matY = (2 * matI32 + 1 - N) / N
    matX = (2 * matK32 + 1 - N) / N
    matR = np.sqrt(matX * matX + matY * matY)
    matT = np.arctan2(matY, matX)
    matV = np.exp(2 * np.pi * matN32 * matR * matR * 1j + matM32 * matT * 1j)
    return matV

def W(matN, matM, matI, matK):
    matN32 = np.float32(matN)
    matM32 = np.float32(matM)
    matI32 = np.float32(matI)
    matK32 = np.float32(matK)
    matY = (2 * matI32 + 1 - N) / N
    matX = (2 * matK32 + 1 - N) / N
    matR = np.sqrt(matX * matX + matY * matY)
    matC = matR <= 1
    matT = np.arctan2(matY, matX)
    matW = np.exp(-2 * np.pi * matN32 * matR * matR * 1j - matM32 * matT * 1j)
    return matW * matC
```

```python
def createMatW():
    matN = np.ones((P,P))
    matM = np.ones((P,P))
    for i in range(P):
        matN[i,:] = i
        matM[:,i] = i
    matN = matN.reshape((P*P,1))
    matN = np.repeat(matN, N*N, axis=1)
    matM = matM.reshape((P*P,1))
    matM = np.repeat(matM, N*N, axis=1)

    matI = np.ones((N,N))
    matK = np.ones((N,N))
    for i in range(N):
        matI[i,:] = i
        matK[:,i] = i
    matI = matI.reshape((1,N*N))
    matI = np.repeat(matI, P*P, axis=0)
    matK = matK.reshape((1,N*N))
    matK = np.repeat(matK, P*P, axis=0)
    arrMatW = []
    for b in range(B):
        s = b * DP
        e = s + DP
        tempW = 4 / (np.pi * N * N) * W(matN[s:e, :], matM[s:e, :], matI[s:e, :], matK[s:e, :])
        arrMatW.append(tempW)
    return np.array(arrMatW)

def createMatV():
    matN = np.ones((P,P))
    matM = np.ones((P,P))
    for i in range(P):
        matN[i,:] = i
        matM[:,i] = i
    matN = matN.reshape((1,P*P))
    matN = np.repeat(matN, N*N, axis=0)
    matM = matM.reshape((1,P*P))
    matM = np.repeat(matM, N*N, axis=0)

    matI = np.ones((N,N))
    matK = np.ones((N,N))
    for i in range(N):
        matI[i,:] = i
        matK[:,i] = i
    matI = matI.reshape((N*N,1))
    matI = np.repeat(matI, P*P, axis=1)
    matK = matK.reshape((N*N,1))
    matK = np.repeat(matK, P*P, axis=1)
    arrMatV = []
    for b in range(B):
        s = b * DN
        e = s + DN
        tempV = V(matN[s:e,:], matM[s:e,:], matI[s:e,:], matK[s:e,:])
        arrMatV.append(tempV)
    return np.array(arrMatV)
```

```python
matW = createMatW()
```

```python
transformedArr = []
for b in range(B):
    tempTr = np.matmul(matW[b], imvec)
    transformedArr.append(tempTr)
transformedArr = np.array(transformedArr)
transformed = transformedArr.reshape((P*P,1))
transformed_abs = np.abs(transformed)
transformed_abs_img = np.reshape(transformed_abs, (P, P))
imshow(transformed_abs_img)
```

```python
matV = createMatV()
```

```python
inversedArr = []
for b in range(B):
    tempIn = np.matmul(matV[b], transformed)
    inversedArr.append(tempIn)
inversedArr = np.array(inversedArr)
inversed = inversedArr.reshape((N*N,1))
inversed_abs = np.abs(inversed)
inversed_abs_img = np.reshape(inversed_abs, (N,N))
imshow(inversed_abs_img)
```
