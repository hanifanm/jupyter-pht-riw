```python
def imshow(imgdata, title=''):
    imgdata_abs = np.abs(imgdata)
    plt.figure()
    plt.title(title)
    plt.imshow(imgdata_abs, cmap = 'gray')
    plt.show()
    
def imsave(imgdata, filename, title=''):
    imgdata_abs = np.abs(imgdata)
    plt.figure()
    plt.title(title)
    plt.imshow(imgdata_abs, cmap = 'gray')
    plt.savefig('dist/' + filename+'.png')
```
